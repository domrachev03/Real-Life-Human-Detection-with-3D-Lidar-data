## Recording rosbag

The package [kitti2bag](https://github.com/tomas789/kitti2bag) provides a handy instrumentary to create a rosbag file from Kitti dataset. 

> Note: the installation requires ROS, and the python, where `kitti2bag` would be installed, should have all ROS packages

Run this in the terminal:
```
pip install kitti2bag
kitti2bag -t 2011_09_28 -r 0054 raw_synced ../rosbags
```

Done, now there is a recorded rosbag file in the folder `rosbags`!
