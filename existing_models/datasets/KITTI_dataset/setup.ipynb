# Kitti dataset download & rosbag generation

Prerequisites:
1. Python with [ROS 1](http://wiki.ros.org/noetic/Installation/Ubuntu) installed
2. Linux tool `unzip` installed (`sudo apt install unzip`)

The following code allows to list all the datasets one want to download and automatically convert them to ros bag files. 

```python
datasets = [
    # (date, number of record from website)
    ('2011_09_28', '0016'),
    ('2011_09_28', '0054')
]
# Python environment with ROS installed
python_with_ros_dir = '/usr/bin/python3'
```

```python
!{python_with_ros_dir} -m pip install kitti2bag
```

```python
for date, number in datasets:
    data_file = f"{date}_drive_{number}_sync.zip"
    calib_file = f"{date}_calib.zip"
    !wget {f'https://s3.eu-central-1.amazonaws.com/avg-kitti/raw_data/{date}_drive_{number}/{data_file}'}
    !wget {f'https://s3.eu-central-1.amazonaws.com/avg-kitti/raw_data/{calib_file}'}

    !unzip -qn {data_file} -d data
    !unzip -qn {calib_file} -d data 

    !kitti2bag -t {date} -r {number} raw_synced data/
```
